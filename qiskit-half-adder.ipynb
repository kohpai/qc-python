```python
from qiskit import QuantumCircuit
from qiskit import Aer, transpile
from qiskit.visualization import plot_histogram
```

```python
qasm_sim = Aer.get_backend('qasm_simulator')
```

```python
def half_adder():
    qc = QuantumCircuit(4)
    # 2 XOR gates
    qc.cx(0, 2)
    qc.cx(1, 2)
    # 1 AND gate for carry bit
    # Toffoli gate is a universal gate
    qc.ccx(0, 1, 3)
    return qc.to_gate()
```

```python
add_0_0 = QuantumCircuit(4, 2)
add_0_0.append(half_adder(), [0, 1, 2, 3])
add_0_0.measure([2,3], [0,1])
add_0_0.draw("mpl")
```

```python
results = qasm_sim.run(transpile(add_0_0, qasm_sim)).result()
counts = results.get_counts()
plot_histogram(counts)
```
