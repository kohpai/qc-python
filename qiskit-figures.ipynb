```python
from qiskit import Aer, QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit.visualization import plot_histogram
```

```python
x = QuantumRegister(3, 'x')
y = QuantumRegister(2, 'y')
z = QuantumRegister(1, 'z')

qc = QuantumCircuit(x, y, z)

qc.cx(x[0], y[0])
qc.cx(x[1], y[0])
qc.cx(x[1], y[1])
qc.cx(x[2], y[1])

qc.mcx(y, z)
```

```python
qc.draw('mpl', 2.0)
```

```python
def empty_gate():
    q = QuantumRegister(4)
    c = QuantumCircuit(q)
    u = c.to_gate()
    u.name = "$V$"
    return u
```

```python
qc.append(empty_gate(), [x[0], x[1], x[2], z[0]])
qc.draw('mpl')
```

```python
qc.mcx(y, z)

qc.cx(x[2], y[1])
qc.cx(x[1], y[1])
qc.cx(x[1], y[0])
qc.cx(x[0], y[0])
```

```python
qc.draw('mpl', 2.0, "compute-uncompute-boolean")
```
